```python
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Iris.csv')
y = df['Species']
x = df.drop(['Id', 'Species'], axis=1)
```

```python
x=df.drop('Id',axis=1)
```

```python
x
```

```python
Xtrain,X_test,y_train,y_test=train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
m_knn=KNeighborsClassifier(n_neighbors=5)
```

```python
# Re-prepare features and target to ensure no string columns in X
y = df['Species']
x = df.drop(['Id', 'Species'], axis=1)

# Re-split the data
Xtrain, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

# Now fit the model
m_knn.fit(Xtrain, y_train)
```

```python
m_knn.score(X_test,y_test)
```

```python
from sklearn.svm import SVC
sv=SVC(C=10,kernel='rbf',gamma='auto')
sv.fit(Xtrain,y_train)
```

```python
sv.score(X_test,y_test)
```

**now using grid search cv**

```python
from sklearn.model_selection import GridSearchCV
classi=GridSearchCV((sv),{
    'C':[1,10,20,30],
    'kernel':['rbf','linear']
},cv=5,return_train_score=False)
```

```python
classi.fit(x,y)
```

```python
classi.cv_results_
```

```python
r=pd.DataFrame(classi.cv_results_)
```

```python
r
```

```python
r[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classi_r=RandomizedSearchCV((sv),{
    'C':[1,10,20,30],
    'kernel':['rbf','linear']
},n_iter=4,cv=5,return_train_score=False)
```

```python
classi_r.fit(x,y)
```

```python
r1=pd.DataFrame(classi_r.cv_results_)
```
