```python
x=df.drop('price',axis=1)
y=df['price']
```

```python
x
```

```python
x.columns
```

```python
x_e=pd.get_dummies(x,columns=['transmission', 'fuelType', 'model'],drop_first=True)
```

```python
x_e
```

```python
x_e=x_e.astype(int)
```

```python
x_e
```

```python
from sklearn.preprocessing import LabelEncoder
col=['model','transmission','fuelType']
xlabel=x.copy()
label_encoders={}
for i in col:
    le=LabelEncoder()
    xlabel[i]=le.fit_transform(xlabel[i].astype(str))
    label_encoders[i]=le
```

```python
xlabel
```

```python
xlabel['model'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
n_c=['year','mileage','tax','mpg']
sc=StandardScaler()
x_e[n_c]=sc.fit_transform(x_e[n_c])
```

```python
x_e
```

```python
xlabel[['model','year','transmission','mileage','fuelType','tax','mpg','engineSize']]=sc.fit_transform(xlabel[['model','year','transmission','mileage','fuelType','tax','mpg','engineSize']])
```

```python
xlabel
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_e, y, test_size=0.33, random_state=42)
```

```python
m=LinearRegression()
m.fit(X_train,y_train)
```

```python
y_p=m.predict(X_test)
```

```python
y_p
```

```python
r2=r2_score(y_test,y_p)
r2
```

```python
n=X_test.shape[0]
p=X_test.shape[1]
a_r=1-((1-r2)*(n-1))/(n-p-1)
a_r
```
