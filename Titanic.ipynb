```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.drop(['Cabin'], axis=1, inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df['Sex']=le.fit_transform(df['Sex'])
df['Embarked']=le.fit_transform(df['Embarked'])
```

```python
df['Name'].dropna(inplace=True)
```

```python
df.drop(['Name'],axis=1,inplace=True)
df.drop(['Ticket'],axis=1,inplace=True)
```

```python
df
```

```python
df=df.astype(int)
```

```python
df
```

```python
x=df.drop(['Survived'],axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_p=model.predict(X_test)
```

```python
y_p
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test,y_p)
```

```python
confusion_matrix(y_test,y_p)
```

```python
classification_report(y_test,y_p)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_sc=sc.fit_transform(X_train)
xt_sc=sc.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_sc,y_train)
```

```python
y_pr_knn=knn.predict(xt_sc)
```

```python
accuracy_score(y_test,y_pr_knn)
```

```python
confusion_matrix(y_test,y_pr_knn)
```

```python
classification_report(y_test,y_pr_knn)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
```

```python
y_pr_nb=nb.predict(X_test)
y_pr_nb
```

```python
accuracy_score(y_test,y_pr_nb)
```

```python
confusion_matrix(y_test,y_pr_nb)
```

```python
classification_report(y_test,y_pr_nb)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=42)
dt.fit(x_sc,y_train)
```

```python
y_pr_dt=dt.predict(xt_sc)
```

```python
accuracy_score(y_test,y_pr_dt)
```

```python
confusion_matrix(y_test,y_pr_dt)
```

```python
classification_report(y_test,y_pr_dt)
```

```python
from sklearn.svm import SVC
svm=SVC(kernel='rbf')
svm.fit(x_sc,y_train)
```

```python
y_pr_sv=svm.predict(xt_sc)
```

```python
y_pr_sv
```

```python
accuracy_score(y_test,y_pr_sv)
```

```python
confusion_matrix(y_test,y_pr_sv)
```

```python
classification_report(y_test,y_pr_sv)
```
